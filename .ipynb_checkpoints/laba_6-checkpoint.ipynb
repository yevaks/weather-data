### TASK_0 

(Cleaning data)

```python
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
```

```python
path = "weather.csv"
df = pd.read_csv(path, names=('d', 'cloudiness(%)', 'precipitation (m.m.)',
                              'D-t.a. temp. (°C)', 'N.a. temp.(°C)',
                  'Wind pow.(m/sec)', 'period'), header=0)

df.replace('-', np.nan, inplace=True)
```

```python
df['cloudiness(%)'] = df['cloudiness(%)'].str.replace('%', '', regex=False).astype(int)

df['precipitation (m.m.)'] = df['precipitation (m.m.)'].str.replace('м.м.', '', regex=False)
df['precipitation (m.m.)'] = pd.to_numeric(df['precipitation (m.m.)'], errors='coerce', downcast='unsigned')
df['precipitation (m.m.)'] =df['precipitation (m.m.)'].astype(pd.Int64Dtype())

df['D-t.a. temp. (°C)'] = df['D-t.a. temp. (°C)'].astype(str)
df['D-t.a. temp. (°C)'] = df['D-t.a. temp. (°C)'].str.replace('°C', '').astype(float)

df['N.a. temp.(°C)'] = df['N.a. temp.(°C)'].astype(str)
df['N.a. temp.(°C)'] = df['N.a. temp.(°C)'].str.replace('°C', '').astype(float)

df['Wind pow.(m/sec)'] = df['Wind pow.(m/sec)'].str.replace('м/с', '', regex=False).astype(int)

```

```python
df['period'] = pd.to_datetime(df['period'])
df['month'] = df['period'].dt.month
df.groupby('month')
df
```

### TASK_1

For period P, plot line graphs of changes in daytime temperature, night temperature, cloudiness, and wind strength by day


```python
P =(df['month'] == 6)
```

```python
gr1=df.loc[P, ['cloudiness(%)', 'D-t.a. temp. (°C)', 'N.a. temp.(°C)', 'Wind pow.(m/sec)']]
gr1
```

```python
gr1.plot(kind='line', subplots=True, layout=(2, 2), figsize=(10, 8))
plt.tight_layout()
```

### TASK_2

The bubble (scatter) graph of daily temperature

```python
data = df[P]

x = data['d']
y = data['D-t.a. temp. (°C)']
sizes = data['precipitation (m.m.)'].fillna(1).size

plt.figure(figsize=(5, 3.5))
plt.scatter(x, y, s=sizes, alpha=0.75, edgecolor='pink', linewidth=2)

plt.xlabel('Day')
plt.ylabel('D-t.a. temp. (°C)')
plt.title('Temperature vs Precipitation')
#sizes = [precipitation[i] if precipitation[i] > 0 else 10 for i in range(len(precipitation))]
```

### TASK_3

Find the monthly average deviation of night temperature from daytime


```python
df['deviation'] = df['D-t.a. temp. (°C)'] - df['N.a. temp.(°C)']

monthly_avg_deviation = df.groupby('month')['deviation'].mean()
monthly_avg_deviation
```

### TASK_4

 The biggest discrepancy between day and night temperatures

```python
larg_values = df.groupby('month', group_keys=False)['deviation'].apply(lambda x: x.abs()).max()
largest_deviation = df.loc[df['deviation'].abs() == larg_values]
largest_deviation
```

### TASK_5

4 windiest days in period P

```python
windiest_days = df[P].nlargest(4, 'Wind pow.(m/sec)')
windiest_days
```

### TASK_6

The number of days with precipitation in the month

```python
precip_days = df[df['precipitation (m.m.)'].notna()].groupby('month')['d'].count()

plt.pie(precip_days, labels=precip_days.index, autopct='%1.1f%%')
plt.title('Days with Precipitation by Month')
```

### TASK_7

Сloudiness

```python
def day_type(cloudiness):
    if cloudiness > 70:
        return 'Cloudy'
    elif cloudiness < 35:
        return 'Sunny'
    else:
        return 'Variable'

df['day_cl'] = df['cloudiness(%)'].apply(day_type)

grouped_data = df.groupby(['month', 'day_cl']).size().unstack(fill_value=0)

grouped_data.plot(kind='bar', stacked=True)

plt.xlabel('Month')
plt.ylabel('Number of Days')
plt.title('Day Type')
```

### TASK_8

All the months in which there were more sunny days than days with precipitation

```python
n_sunny = df.groupby('month')['cloudiness(%)'].apply(lambda x: (x < 35).sum())
n_perc = df.groupby('month')['precipitation (m.m.)'].count()

sunny_months = n_sunny[n_sunny > n_perc].index
print('Lucky sunny days:)')
print(*sunny_months, sep=',\n')
```
